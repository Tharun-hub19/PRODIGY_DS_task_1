```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set a seed so the "random" data is the same every time you run it
np.random.seed(42)

# Generate synthetic population data
n = 1000
df = pd.DataFrame({
    'Age': np.random.randint(0, 81, size=n),
    'Gender': np.random.choice(['Male', 'Female', 'Other'], size=n, p=[0.48, 0.49, 0.03])
})

# Display the first 5 rows to make sure it looks correct
df.head()
```

```python
plt.figure(figsize=(10, 5))

# kde=True adds the "smooth line" over the bars
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue')

plt.title('Age Distribution of the Population')
plt.xlabel('Age (Years)')
plt.ylabel('Number of People')
plt.show()
```

```python
# Count occurrences of each gender
gender_counts = df['Gender'].value_counts().reset_index()
gender_counts.columns = ['Gender', 'Count']

# Sort them so the tallest bar is on the left
gender_counts = gender_counts.sort_values(by='Count', ascending=False)

gender_counts
```

```python
plt.figure(figsize=(8, 5))

# Fixing the warning: map x to hue and set legend=False
sns.barplot(
    data=gender_counts, 
    x='Gender', 
    y='Count', 
    hue='Gender', 
    palette='viridis', 
    legend=False
)

plt.title('Gender Distribution of the Population')
plt.ylabel('Total Count')
plt.show()
```
